## Q1. Partition String 
`Medium`, `4 pt.`

Given a string `s`, partition it into unique segments according to the following procedure:

- Start building a segment beginning at index 0.
- Continue extending the current segment character by character until the current segment has not been seen before.
- Once the segment is unique, add it to your list of segments, mark it as seen, and begin a new segment from the next index.
- Repeat until you reach the end of `s`.

Return an array of strings `segments`, where `segments[i]` is the `ith` segment created.

```python
# Original Solution -> TIME LIMIT EXCEEDED

class Solution(object):
    def partitionString(self, s):
        output = []
        cindex = 0
        while cindex < len(s):
            a = s[cindex]
            while 1:
                if a in output:
                    try:
                        a = a + s[cindex+1]
                        cindex += 1
                    except:
                        cindex += 1
                        break
                else:
                    output.append(a)
                    cindex += 1
                    break
        return output
```

## Q2. Longest Common Prefix Between Adjacent Strings After Removals
`Medium`, `4 pt.`

You are given an array of strings `words`. For each index `i` in the range `[0, words.length - 1]`, perform the following steps:

- Remove the element at index `i` from the `words` array.
- Compute the **length** of the **longest common prefix** among all **adjacent** pairs in the modified array.
  
Return an array `answer`, where `answer[i]` is the length of the longest common prefix between the adjacent pairs after removing the element at index `i`. If no adjacent pairs remain or if none share a common prefix, then `answer[i]` should be 0.

```python
# Original Solution -> WRONG ANSWER

class Solution(object):
    def longestCommonPrefix(self, words):
        output = []
        for i in range(len(words)):
            removed = words[:i] + words[i+1:]
            w = 0
            k = 0
            prefix = ''
            p = ''
            while w < len(removed)-1:
                if (len(removed[w]) == k):
                    w += 1
                    k = 0
                    if len(p) > len(prefix):
                        prefix = p
                        p = ''
                if (w == len(removed)-1):
                    break
                if removed[w][k] == removed[w+1][k]:
                    p = p + removed[w][k]
                    k += 1
                else:
                    w += 1
                    k = 0
                    if len(p) > len(prefix):
                        prefix = p
                        p = ''
            output.append(len(prefix))    
        return output
        
```

## Q3. Partition Array to Minimize XOR
`Medium`, `5 pt.`

You are given an integer array `nums` and an integer `k`.

Your task is to partition `nums` into `k` non-empty subarrays. For each subarray, compute the bitwise **XOR** of all its elements.

Return the **minimum** possible value of the **maximum XOR** among these k subarrays.

```python
# Original Solution -> TIME LIMIT EXCEEDED

class Solution(object):
    def minXor(self, nums, k):
        max_xor_list = []
        def maxxorlist(s, xor_list, left_partitions):
            if left_partitions == 1:
                part = nums[s:]
                xorv = part[0]
                if len(part) > 1:
                    for k in part[1:]:
                        xorv = xorv ^ k
                xor_list = xor_list + [xorv]
                max_xor_list.append(max(xor_list))
                return

            for i in range(s + 1, len(nums) - left_partitions + 2):
                part = nums[s:i]
                xorv = part[0]
                if len(part) > 1:
                    for k in part[1:]:
                        xorv = xorv ^ k
                maxxorlist(i, xor_list + [xorv], left_partitions - 1)
        
        maxxorlist(0, [], k)
        output = min(max_xor_list)
        return output
    
```

## Q4. Maximize Spanning Tree Stability with Upgrades
`Hard`, `6 pt.`

You are given an integer `n`, representing `n` nodes numbered from 0 to `n - 1` and a list of `edges`, where `edges[i] = [ui, vi, si, musti]`:

- `u_i` and `v_i` indicates an undirected edge between nodes `u_i` and `v_i`.
- `s_i` is the strength of the edge.
- `must_i` is an integer (0 or 1). If `must_i == 1`, the edge **must** be included in the **spanning tree**. These edges **cannot** be **upgraded**.

You are also given an integer `k`, the **maximum** number of upgrades you can perform. Each upgrade **doubles** the strength of an edge, and each eligible edge (with `must_i == 0`) can be upgraded **at most** once.

The **stability** of a spanning tree is defined as the **minimum** strength score among all edges included in it.

Return the **maximum** possible stability of any valid spanning tree. If it is impossible to connect all nodes, return `-1`.

**Note**: A **spanning tree** of a graph with `n` nodes is a subset of the edges that connects all nodes together (i.e. the graph is **connected**) without forming any cycles, and uses **exactly** `n - 1` edges.

 

```python
# Original Solution -> WRONG ANSWER

class Solution(object):
    def maxStability(self, n, edges, k):
        strength_list = []
        upgradable = []
        for e in edges:
            if e[3] == 1:
                strength_list.append(e[2])
            else:
                upgradable.append(e[2])
        if len(upgradable) == 0:
            output = -1
        else:
            max_k_upgradable = sorted(upgradable, reverse=True)[:k]
            max_k_upgraded = [i * 2 for i in max_k_upgradable]
            strength_list = strength_list + max_k_upgraded
            output = min(strength_list)
        return output
```
