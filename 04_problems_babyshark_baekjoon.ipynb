# 아기 상어 [🔗](https://www.acmicpc.net/problem/16236)

## 문제
N×N 크기의 공간에 물고기 M마리와 아기 상어 1마리가 있다. 공간은 1×1 크기의 정사각형 칸으로 나누어져 있다. 한 칸에는 물고기가 최대 1마리 존재한다.

아기 상어와 물고기는 모두 크기를 가지고 있고, 이 크기는 자연수이다. 가장 처음에 아기 상어의 크기는 2이고, 아기 상어는 1초에 상하좌우로 인접한 한 칸씩 이동한다.

아기 상어는 자신의 크기보다 큰 물고기가 있는 칸은 지나갈 수 없고, 나머지 칸은 모두 지나갈 수 있다. 아기 상어는 자신의 크기보다 작은 물고기만 먹을 수 있다. 따라서, 크기가 같은 물고기는 먹을 수 없지만, 그 물고기가 있는 칸은 지나갈 수 있다.

아기 상어가 어디로 이동할지 결정하는 방법은 아래와 같다.

더 이상 먹을 수 있는 물고기가 공간에 없다면 아기 상어는 엄마 상어에게 도움을 요청한다.
먹을 수 있는 물고기가 1마리라면, 그 물고기를 먹으러 간다.
먹을 수 있는 물고기가 1마리보다 많다면, 거리가 가장 가까운 물고기를 먹으러 간다.
거리는 아기 상어가 있는 칸에서 물고기가 있는 칸으로 이동할 때, 지나야하는 칸의 개수의 최솟값이다.
거리가 가까운 물고기가 많다면, 가장 위에 있는 물고기, 그러한 물고기가 여러마리라면, 가장 왼쪽에 있는 물고기를 먹는다.
아기 상어의 이동은 1초 걸리고, 물고기를 먹는데 걸리는 시간은 없다고 가정한다. 즉, 아기 상어가 먹을 수 있는 물고기가 있는 칸으로 이동했다면, 이동과 동시에 물고기를 먹는다. 물고기를 먹으면, 그 칸은 빈 칸이 된다.

아기 상어는 자신의 크기와 같은 수의 물고기를 먹을 때 마다 크기가 1 증가한다. 예를 들어, 크기가 2인 아기 상어는 물고기를 2마리 먹으면 크기가 3이 된다.

공간의 상태가 주어졌을 때, 아기 상어가 몇 초 동안 엄마 상어에게 도움을 요청하지 않고 물고기를 잡아먹을 수 있는지 구하는 프로그램을 작성하시오.

## 입력
첫째 줄에 공간의 크기 N(2 ≤ N ≤ 20)이 주어진다.

둘째 줄부터 N개의 줄에 공간의 상태가 주어진다. 공간의 상태는 0, 1, 2, 3, 4, 5, 6, 9로 이루어져 있고, 아래와 같은 의미를 가진다.

- 0: 빈 칸
- 1, 2, 3, 4, 5, 6: 칸에 있는 물고기의 크기
- 9: 아기 상어의 위치

아기 상어는 공간에 한 마리 있다.

## 출력
첫째 줄에 아기 상어가 엄마 상어에게 도움을 요청하지 않고 물고기를 잡아먹을 수 있는 시간을 출력한다.

```python
# CAN'T USE PANDAS...:(
import pandas as pd
```

```python
# Searching way to go up/left/right/down
def search_nearby(loc_r, loc_c, prev_pos, prev_sec):
    path = pd.DataFrame({'pos':[], 'sec':[]}) # possible paths
    eat = pd.DataFrame({'pos':[], 'sec':[]}) # fishes to eat
    for m in [(-1, 0), (0, -1), (0, 1), (1, 0)]: # search up/left/right/down
        new_pos = prev_pos.copy() # new position
        if [loc_r + m[0], loc_c + m[1]] in prev_pos: # if already visited
            continue
        elif (loc_r + m[0] == -1) | (loc_c + m[1] == -1) | (loc_r + m[0] >= N) | (loc_c + m[1] >= N): # if out of boundary
            continue
        elif (S[loc_r + m[0]][loc_c + m[1]] > 0) & (S[loc_r + m[0]][loc_c + m[1]] < b): # if there's a fish that the baby can eat
            r = loc_r + m[0]
            c = loc_c + m[1]
            new_pos.append([r, c])
            new = pd.DataFrame({'pos':[new_pos], 'sec':[prev_sec + 1]})
            path = pd.concat([path, new], ignore_index=True)
            eat = pd.concat([eat, new], ignore_index=True)
            break
        elif (S[loc_r + m[0]][loc_c + m[1]] == 0) | (S[loc_r + m[0]][loc_c + m[1]] == b): # if there's no fish or a fish of the same size with the baby
            r = loc_r + m[0]
            c = loc_c + m[1]
            new_pos.append([r, c])
            new = pd.DataFrame({'pos':[new_pos], 'sec':[prev_sec + 1]})
            path = pd.concat([path, new], ignore_index=True)
        else: # others(if a fish is too large)
            continue

    return path, eat

# Get input
N = int(input())
S = [list(map(int, input().split())) for _ in range(N)]

# Current location of the baby shark
loc_r = 0
loc_c = 0

for i, l in enumerate(S):
    try:
        loc_c = l.index(9)
        loc_r = i
    except:
        continue

# Basic variables
f = 0 # number of fish eaten
b = 2 # size of the baby shark
total_sec = 0 # total time the baby shark spent

while 1:
    # Ask mama for help when
    unique_S = list(set([i for l in S for i in l])) # unique values in the space
    if (0 in unique_S) & (sum(map(lambda x: x < b, unique_S[1:])) == 0): # no more fish left to eat
        break
    elif (len(unique_S) == 2): # ate up all
        break
    elif (0 not in unique_S) & (sum(map(lambda x: x < b, unique_S)) == 0): # no fish smaller than the baby
        break

    # Search distance=1
    path, eat = search_nearby(loc_r, loc_c, [[loc_r, loc_c]], 0) 

    # Search more
    if path.empty: # if the baby's stuck
        break
    elif eat.empty: # if the baby didn't find a fish to eat yet
        while 1:
            empty_temp = 0 # count 
            new_path = pd.DataFrame({'pos':[], 'sec':[]}) # possible paths on current phase
            eat = pd.DataFrame({'pos':[], 'sec':[]}) # paths that ends at a fish's location
            for i in range(len(path)):
                # last location of the baby
                n_loc_r = path.loc[i, 'pos'][-1][0] 
                n_loc_c = path.loc[i, 'pos'][-1][1]
                
                # find possible paths from the last location
                temp_path, temp_eat = search_nearby(n_loc_r, n_loc_c, path.loc[i, 'pos'], path.loc[i, 'sec'])

                if temp_path.empty: # if stuck
                    empty_temp += 1
                    
                if temp_eat.empty: # if not stuck, and nothing to eat
                    new_path = pd.concat([new_path, temp_path], ignore_index=True) 
                else: # found a fish to eat
                    eat = pd.concat([eat, temp_eat], ignore_index=True) 
                    
            if empty_temp == len(new_path): # no way to go
                break            
            elif eat.empty: # there's still ways to go, but nothing to eat
                path = new_path 
            else: # found a fish to eat
                break
                
    if eat.empty: # haven't found a fish to eat
        continue
    else:  
        # find a fish that's on the highest, most left position
        last = eat['pos'].map(lambda x: x[-1]) 
        if all(l == last[0] for l in last): # if there's just one target found
            target = eat.loc[0]
        else:
            target_idx = eat['pos'].map(lambda x: x[-1]).sort_values().index[0]
            target = eat.loc[target_idx]

        # move the baby
        S[loc_r][loc_c] = 0
        loc_r = target['pos'][-1][0]
        loc_c = target['pos'][-1][1]
        S[loc_r][loc_c] = 9
        
        f += 1
        total_sec += target['sec']
        if f == b:
            f = 0
            b += 1

print(total_sec)
```
